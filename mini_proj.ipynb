```python
import numpy as np
import scipy as sp
```

```python
a = [ [np.random.normal() for i in range(3)]    for j in range(4) ]
a = np.array(a)
A = a.T @ a
print(A)
```

```python
b = [ [np.random.normal() for i in range(3)]    for j in range(4) ]
b = np.array(b)
B = b.T @ b
print(B)

# print(np.linalg.det(sp.linalg.khatri_rao(A,B))/(np.linalg.det(A) * np.linalg.det(B)))
```

```python
Y = []
n = 100
for i in range(n):
  # b = [ [np.random.normal() for i in range(3)]    for j in range(4) ]
  # b = np.array(b)
  # B = b.T @ b
  b = np.copy(a)
  p = np.random.randint(0,4)
  q = np.random.randint(0,3)
  b[p,q] = b[p,q]*0.1

  p = np.random.randint(0,4)
  q = np.random.randint(0,3)
  b[p,q] = b[p,q]*0.1

  B = b.T @ b
  det_AB = np.linalg.det(A*B)     # det(A*B)
  Y.append(det_AB/(np.linalg.det(A) * np.linalg.det(B)))

print("Maximum = ", max(Y))
print("Minimum = ", min(Y))
X = np.arange(1,n+1)
Y = np.array(Y)

import matplotlib.pyplot as plt
plt.title("Line graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(X, Y, color ="green")
plt.show()
```
